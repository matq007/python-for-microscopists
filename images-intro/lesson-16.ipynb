# Lesson 16 - Intro to image

```python
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
```

```python
image = io.imread('image.jpg')
plt.imshow(image)
```

```python
print(f'Size: {image.shape}') # (width, height, R/G/B)
print(f'Type: {image.dtype}')
```

```python
random_image = np.random.random([300, 300])
print(f'Minimum: {random_image.min()}\nMaximum: {random_image.max()}')
plt.imshow(random_image)
```

```python
# Modified image
plt.imshow(image * 47)
```

```python
# width: 200, height: 200
image[10:200, 10:200, :] = [255, 0, 0]
plt.imshow(image)
```
