## Skicit-image

```python
from skimage import io, img_as_float, img_as_ubyte
```

```python
img = io.imread("image.jpg")
print(f'Min: {img.min()}\nMax: {img.max()}')
print(f'Type: {type(img)}\nnumpy type: {img.dtype}\nshape: {img.shape}')
```

Image can be converted to different data type. Here we convert the image to float.

**Warning**
- this function `img_as_float` loads and scales the image (0, 1)
- if you just want to convert the image values to float you can use `img.astype(np.float)` no scaling just converting `int` to `float`

```python
img_float = img_as_float(img)
print(f'Min: {img_float.min()}\nMax: {img_float.max()}')
print(f'Type: {type(img_float)}\nnumpy type: {img_float.dtype}\nshape: {img_float.shape}')
```

```python
plt.imshow(img)
```

## OpenCV

Perfect for motion analysis

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# OpenCV represents RGB colors in reverse order => BGR
img = cv2.imread("image.jpg")
plt.imshow(img)
plt.title('Default OpenCV')
plt.show()

# original colors
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original colors')
plt.show()

# To open the image as grey use img = cv2.imread("image.jpg", 0)
img_gray = cv2.imread("image.jpg", 0)
plt.title('Grey colors')
plt.imshow(img_gray, cmap='gray')
```

## Read multiple images

```python
import cv2
import glob

for file in glob.glob('*.jpg'):
    img = cv2.imread(file)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```
